```python
import seaborn as sns

top = df['artist_top_genre'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=top[:5].index,y=top[:5].values)
plt.xticks(rotation=45)
plt.title('Top genres',color = 'blue')
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
sns.set_theme(style="ticks")

data = df[df['artist_top_genre'].str.contains('afro dancehall', case=True, regex=True)]

# Show the joint distribution using kernel density estimation
g = sns.jointplot(
    data=data,
    x="popularity", y="danceability", hue="artist_top_genre",
    kind="kde",
)
```

```python
df.plot(kind="scatter", x="danceability", y="acousticness")
```

```python
import seaborn as sns

sns.FacetGrid(df, hue="artist_top_genre", size=5) \
   .map(plt.scatter, "danceability", "acousticness") \
   .add_legend()
```

```python
sns.swarmplot(x=df['danceability'],
              y=df['artist_top_genre'])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# Calculate the linkage: mergings
mergings = linkage(df,method='complete')


# Plot the dendrogram, using names as labels
dendrogram(mergings)
plt.title('Hierarchical Clustering')
plt.figure(figsize=(200, 60))
plt.show()
```

```python
df = pd.read_csv("../../data/nigerian-songs.csv")
new_columns = ['acousticness', 'danceability']
df = df.drop([c for c in df.columns if c not in new_columns], axis=1)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(df)
labels = kmeans.predict(df)
plt.scatter(df['acousticness'],df['danceability'],c = labels)
plt.xlabel('danceability')
plt.xlabel('acousticness')
plt.show()
```
